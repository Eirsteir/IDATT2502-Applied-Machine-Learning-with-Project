```python
import torch
import matplotlib.pyplot as plt

from sigmoid_model import SigmoidModel

# NOT: x is inverted
train_x = torch.tensor([[1.0], [0.0]]).reshape(-1,1) # input
train_y = torch.tensor([[0.0], [1.0]]).reshape(-1,1) # output
```

```python
model = SigmoidModel()

optimizer = torch.optim.SGD([model.b, model.W], lr=0.1)
for epoch in range(250_000):
	model.loss(train_x, train_y).backward()
	optimizer.step()
	optimizer.zero_grad()

print("W = %s, b = %s, loss = %s" %(model.W, model.b, model.loss(train_x, train_y)))
```

```python

plt.plot(train_x, train_y, 'o', label='$(x^{(i)},y^{(i)})$')
plt.xlabel('x')
plt.ylabel('y')
x = torch.arange(0.0,1.0, 0.01).reshape(-1,1)
plt.plot(x, model.f(x).detach(), label='$\\hat y = f(x) = \sigma(xW + b)$')
plt.legend()
plt.show()
```
